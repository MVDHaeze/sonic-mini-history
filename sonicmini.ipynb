```python
import json
import pandas as pd
import glob
```

```python
# load all json from subfolder in historyExport into a pandas dataframe
final_paths = []
extract_path= './historyExport'
files = glob.glob(extract_path+ '/**/*.json', recursive=True)

for file in files:
    with open(file, 'r') as f:
        json_file = json.load(f)
        output_path = json_file
        final_paths.append(output_path)

print(final_paths)

df = pd.DataFrame(final_paths)
```

```python
# Keep only the relevant column for computing total number hours of printing
df_light = df[['name','printerId','remindDate','timeEstimated']].copy()
```

```python
# Convert timeEstimated (in milliseconds) to timeEstimatedinH (in hours)
df_light["timeEstimatedinH"]=df_light["timeEstimated"]/3600000
# Convert remindDate (in xx) to readable timestamp
df_light.drop("timeEstimated", axis=1)
pd.options.display.float_format = '{:,.2f}'.format
#df_light["timeEstimatedinH"]
```

```python
df[['printerId','timeEstimatedinH']].to_csv("sonic mini.csv")
```
