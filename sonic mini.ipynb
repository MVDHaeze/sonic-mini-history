```python
import json
import os, shutil
import pandas as pd
import glob
import math
```

```python
final_paths = []
extract_path= 'C:/Users/Maxime/AppData/Roaming/LycheeSlicer/historyExport'
files = glob.glob(extract_path+ '/**/*.json', recursive=True)

for file in files:
    with open(file, 'r') as f:
        json_file = json.load(f)
        output_path = json_file
        final_paths.append(output_path)

print(final_paths)
```

```python
df = pd.DataFrame(final_paths)
```

```python
df["timeEstimatedinH"]=df["timeEstimated"]/3600000
pd.options.display.float_format = '{:,.2f}'.format
df["timeEstimatedinH"]
```
